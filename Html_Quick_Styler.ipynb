```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_name = "ibm-granite/granite-3.3-2b-instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float16,
    device_map="auto"
)

print("Model Loaded Successfully!")
```

```python
class ColorPaletteGenerator:
    def __init__(self):
        self.palettes = {
            "modern": {
                "primary": "#4F46E5",
                "secondary": "#6366F1",
                "accent": "#22D3EE",
                "background": "#0F172A",
                "surface": "#1E293B",
                "text_primary": "#F8FAFC",
                "text_secondary": "#CBD5E1",
                "border": "#334155"
            },
            "minimal": {
                "primary": "#000000",
                "secondary": "#333333",
                "accent": "#777777",
                "background": "#FFFFFF",
                "surface": "#F5F5F5",
                "text_primary": "#111111",
                "text_secondary": "#555555",
                "border": "#DDDDDD"
            }
        }

    def get_palette(self, style):
        return self.palettes.get(style, self.palettes["modern"])
```

```python
class HTMLPageBuilder:
    def __init__(self, palette):
        self.palette = palette

    def generate_html(self, title, description):
        html = f"""
        <!DOCTYPE html>
        <html>
        <head>
            <title>{title}</title>
            <style>
                body {{
                    background-color: {self.palette['background']};
                    color: {self.palette['text_primary']};
                    font-family: Arial, sans-serif;
                }}
                header {{
                    background-color: {self.palette['primary']};
                    padding: 20px;
                    text-align: center;
                }}
                section {{
                    padding: 40px;
                }}
                footer {{
                    background-color: {self.palette['surface']};
                    padding: 20px;
                    text-align: center;
                }}
            </style>
        </head>
        <body>
            <header>
                <h1>{title}</h1>
            </header>
            <section>
                <p>{description}</p>
            </section>
            <footer>
                <p>© 2026 {title}</p>
            </footer>
        </body>
        </html>
        """
        return html
```

```python
import gradio as gr

def generate_page(title, description, style):
    palette_gen = ColorPaletteGenerator()
    palette = palette_gen.get_palette(style)

    builder = HTMLPageBuilder(palette)
    html = builder.generate_html(title, description)
    return html

with gr.Blocks() as app:
    gr.Markdown("# HTML Quick-Styler")

    with gr.Tab("Generate Page"):
        title = gr.Textbox(label="Page Title")
        description = gr.Textbox(label="Description")
        style = gr.Dropdown(["modern", "minimal"], label="Color Style")
        output = gr.HTML()

        btn = gr.Button("Generate")
        btn.click(generate_page, inputs=[title, description, style], outputs=output)

app.launch(share=True)
```
